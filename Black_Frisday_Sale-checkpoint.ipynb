## Data Preprocessing

### Creating Dummy Variables for Categorical Features

```python
data.head()
```

```python
new_data = pd.get_dummies(data, columns=['Gender','Age','City_Category','Stay_In_Current_City_Years'],
                               drop_first=True)
```

### Removing unwanted Columns

```python
# as about 70% of data is missing fromProduct_Category_3 col
# its better to drop it

new_data.drop(labels=['User_ID', 'Product_ID', 'Product_Category_3'], axis=1, inplace=True)
print("Columns User_ID, Product_ID and Product_Category_3 dropped succesfully")
```

```python
new_data.head()
```

### Filling Missing Values

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
```

```python
new_data['Product_Category_2'] = imputer.fit_transform(data['Product_Category_2'].values.reshape(-1,1))
```

```python
new_data.head()
```

## Exploratory Data Analysis

```python
#Purchase mean for male purchasers
MalePurchaseData = data[data['Gender'] == 'M']
MalePurchaseMean = np.mean(MalePurchaseData['Purchase'])

#Purchase mean for female purchasers
FemalePurchaseData = data[data['Gender'] == 'F']
FemalePurchaseMean = np.mean(FemalePurchaseData['Purchase'])

# Visualising Means of Male vs Female on Purchase column
labels = ['Males', 'Females']
values = [MalePurchaseMean,FemalePurchaseMean]

plt.bar(labels, values, facecolor='g', edgecolor='black', alpha=0.4)
```

```python
genderCount = [MalePurchaseData.shape[0], FemalePurchaseData.shape[0]]


plt.pie(genderCount,labels=labels,labeldistance=1.1,pctdistance=0.4,autopct='%1.1f%%', shadow=True)
plt.legend(labels)
plt.title('Ratio of Male vs Female Purchasers in Black Friday Sale')
```

```python
labels = []
values = []

for uniqueCity in data['City_Category'].sort_values().unique():
    
    CityPurchaseData = data[data['City_Category'] == uniqueCity]
    CityPurchaseMean = np.mean(CityPurchaseData['Purchase'])
    labels.append(uniqueCity)
    values.append(CityPurchaseMean)
    
plt.bar(labels,values, width=.9, facecolor='g', edgecolor='k', alpha=.5)
plt.text(-0.7,10800,'Average purchase in black friday based on City_Category')
```

```python
values = []

for CityCatCount in data['City_Category'].value_counts():
    values.append(CityCatCount)

plt.pie(values, labels=labels,
              explode=[0.1,0.05,0.05],
              autopct='%1.1f%%',
              shadow=True,
              startangle=0,
              labeldistance=1.1,
              pctdistance=.6)
plt.legend(labels)
```

```python
values = [data['Product_Category_1'].sum(), data['Product_Category_2'].sum()]
labels = ['Product_Category_1','Product_Category_2']

plt.bar(labels,values, width=.9, facecolor='b', edgecolor='k', alpha=.5)
plt.text(-0.7,6000000,'Compar between number of boughts of category_1 and category_2 in black friday')
```

## Train Test Split

```python
X = new_data.iloc[ :,new_data.columns != 'Purchase'].values
y = new_data.Purchase.values

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

## Creating The Model

```python
from catboost import CatBoostRegressor

regressor = CatBoostRegressor()
regressor.fit(X_train,y_train)
```

```python
predictions = regressor.predict(X_test)
```

## Evaluating the Model

```python
from sklearn.metrics import mean_squared_error
print(f'Mean Squared Error : {mean_squared_error(y_test,predictions)}')
print(f'Root Mean Squared Error : {np.sqrt(mean_squared_error(y_test,predictions))}')
```
